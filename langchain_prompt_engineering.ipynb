## ✅ Task 1: Replace Hardcoded Prompts (10%)

**Goal:** Convert a hardcoded f-string function into a reusable LangChain `PromptTemplate`.

**Original (Hardcoded):**
```python
def explain_topic(topic):
    return f"Explain {topic} in simple terms for beginners"
```

````python
# ============================================================
# TASK 1: Replace Hardcoded Prompts with LangChain PromptTemplate
# ============================================================

# Define a reusable PromptTemplate — no f-strings, no hardcoding
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Reusable function using the template
def explain_topic(topic: str) -> str:
    """
    Generates a beginner-friendly explanation prompt for a given topic.
    Uses LangChain PromptTemplate for reusability.
    """
    return explain_template.format(topic=topic)


# --- Test the template with different topics ---
test_topics = ["Machine Learning", "Blockchain", "Neural Networks"]

print("📌 Task 1 Output – Reusable PromptTemplate:\n")
for t in test_topics:
    print(f"  Topic: '{t}'")
    print(f"  Prompt: {explain_topic(t)}")
    print()
````

## ✅ Task 2: Multi-Input Prompt System (15%)

**Goal:** Build a prompt template that accepts multiple inputs — `topic`, `audience`, and `tone`.

**Template:** `Explain {topic} for {audience} in a {tone} tone`

````python
# ============================================================
# TASK 2: Multi-Input Prompt System
# ============================================================

# Define template with multiple input variables
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def multi_input_prompt(topic: str, audience: str, tone: str) -> str:
    """
    Generates a prompt tailored to a specific audience and tone.
    Accepts: topic, audience, tone
    """
    return multi_input_template.format(topic=topic, audience=audience, tone=tone)


# --- Test Cases as specified in the assignment ---
test_cases = [
    ("AI",             "beginners",  "friendly"),
    ("Python",         "kids",        "fun"),
    ("Deep Learning",  "engineers",   "technical"),
]

print("📌 Task 2 Output – Multi-Input Prompt System:\n")
for topic, audience, tone in test_cases:
    prompt = multi_input_prompt(topic, audience, tone)
    print(f"  Inputs  → topic: '{topic}' | audience: '{audience}' | tone: '{tone}'")
    print(f"  Prompt  → {prompt}")
    print()
````

## ✅ Task 3: Prompt Variations Engine (15%)

**Goal:** Create 3 distinct `PromptTemplate`s for different styles and run them for `topic = Machine Learning`.

| Style | Template |
|---|---|
| Teaching | Explain {topic} clearly step by step |
| Interview | Ask 3 questions about {topic} |
| Storytelling | Explain {topic} as a story |

````python
# ============================================================
# TASK 3: Prompt Variations Engine
# ============================================================

# --- Define 3 distinct PromptTemplates ---

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Map style names to their templates for clean lookup
variation_templates = {
    "Teaching":     teaching_template,
    "Interview":    interview_template,
    "Storytelling": storytelling_template,
}

def prompt_variations_engine(topic: str) -> dict:
    """
    Generates a prompt for each variation style (Teaching, Interview, Storytelling)
    for a given topic.
    Returns a dictionary: {style_name: generated_prompt}
    """
    return {style: template.format(topic=topic)
            for style, template in variation_templates.items()}


# --- Run for Machine Learning ---
topic = "Machine Learning"
results = prompt_variations_engine(topic)

print(f"📌 Task 3 Output – Prompt Variations Engine (Topic: '{topic}'):\n")
for style, prompt in results.items():
    print(f"  [{style}]")
    print(f"   → {prompt}")
    print()
````

## ✅ Task 4: ChatPromptTemplate System (15%)

**Goal:** Build a role-based chat prompt system using `ChatPromptTemplate` with `System` and `User` messages.

**Roles:** `teacher`, `interviewer`, `motivator`

````python
# ============================================================
# TASK 4: ChatPromptTemplate System
# ============================================================

# Define system behavior for each role
ROLE_SYSTEM_MESSAGES = {
    "teacher":     "You are a patient and knowledgeable teacher. Explain concepts clearly with examples.",
    "interviewer": "You are a professional interviewer. Ask insightful and relevant interview questions.",
    "motivator":   "You are an enthusiastic motivator. Inspire and encourage learners with energy and positivity.",
}

def build_chat_prompt(role: str, topic: str) -> list:
    """
    Builds a ChatPromptTemplate for the given role and topic.

    Args:
        role  (str): One of 'teacher', 'interviewer', 'motivator'
        topic (str): The subject to discuss

    Returns:
        list: Formatted chat messages (system + user)
    """
    # Validate role
    if role not in ROLE_SYSTEM_MESSAGES:
        raise ValueError(f"Invalid role '{role}'. Choose from: {list(ROLE_SYSTEM_MESSAGES.keys())}")

    # Build the ChatPromptTemplate dynamically
    chat_template = ChatPromptTemplate.from_messages([
        ("system", ROLE_SYSTEM_MESSAGES[role]),      # System: defines AI behavior
        ("human",  "Please discuss the topic: {topic}")  # User: dynamic input
    ])

    # Format and return messages
    return chat_template.format_messages(topic=topic)


# --- Test with role='teacher' and topic='Neural Networks' ---
test_cases = [
    ("teacher",     "Neural Networks"),
    ("interviewer", "Machine Learning"),
    ("motivator",   "Deep Learning"),
]

print("📌 Task 4 Output – ChatPromptTemplate System:\n")
for role, topic in test_cases:
    messages = build_chat_prompt(role, topic)
    print(f"  Role: '{role}' | Topic: '{topic}'")
    for msg in messages:
        msg_type = msg.__class__.__name__.replace('Message', '')
        print(f"    [{msg_type}] → {msg.content}")
    print()
````

## ✅ Task 5: Input Validation Layer (10%)

**Goal:** Create a `validate_inputs(audience, tone)` function that enforces allowed values.

- `audience` ∈ `[beginner, intermediate, expert]`
- `tone` ∈ `[formal, casual, fun]`

````python
# ============================================================
# TASK 5: Input Validation Layer
# ============================================================

# Allowed values for validation
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES     = ["formal", "casual", "fun"]

# Default fallback values
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "casual"


def validate_inputs(audience: str, tone: str, use_defaults: bool = True) -> tuple:
    """
    Validates audience and tone inputs against allowed values.

    Args:
        audience     (str):  Target audience level.
        tone         (str):  Tone of the explanation.
        use_defaults (bool): If True, assign defaults on invalid input.
                             If False, raise ValueError on invalid input.

    Returns:
        tuple: (validated_audience, validated_tone)

    Raises:
        ValueError: If use_defaults=False and input is invalid.
    """
    audience = audience.strip().lower()
    tone     = tone.strip().lower()
    errors   = []

    # Validate audience
    if audience not in VALID_AUDIENCES:
        if use_defaults:
            print(f"  ⚠️  Invalid audience '{audience}'. Defaulting to '{DEFAULT_AUDIENCE}'.")
            audience = DEFAULT_AUDIENCE
        else:
            errors.append(f"Invalid audience '{audience}'. Must be one of: {VALID_AUDIENCES}")

    # Validate tone
    if tone not in VALID_TONES:
        if use_defaults:
            print(f"  ⚠️  Invalid tone '{tone}'. Defaulting to '{DEFAULT_TONE}'.")
            tone = DEFAULT_TONE
        else:
            errors.append(f"Invalid tone '{tone}'. Must be one of: {VALID_TONES}")

    # Raise combined errors if not using defaults
    if errors:
        raise ValueError(" | ".join(errors))

    return audience, tone


# --- Test Cases ---
print("📌 Task 5 Output – Input Validation Layer:\n")

# Valid inputs
print("  Test 1 – Valid inputs:")
result = validate_inputs("expert", "formal")
print(f"  ✅ Result → audience: '{result[0]}', tone: '{result[1]}'\n")

# Invalid audience → fallback to default
print("  Test 2 – Invalid audience (use_defaults=True):")
result = validate_inputs("novice", "fun")
print(f"  ✅ Result → audience: '{result[0]}', tone: '{result[1]}'\n")

# Invalid tone → fallback to default
print("  Test 3 – Invalid tone (use_defaults=True):")
result = validate_inputs("intermediate", "aggressive")
print(f"  ✅ Result → audience: '{result[0]}', tone: '{result[1]}'\n")

# Invalid inputs → raise error
print("  Test 4 – Invalid inputs (use_defaults=False → raises error):")
try:
    validate_inputs("unknown", "weird", use_defaults=False)
except ValueError as e:
    print(f"  ❌ ValueError caught → {e}\n")
````

## ✅ Task 6: Prompt Generator App (15%)

**Goal:** Create a `generate_prompt(topic, audience, tone, style)` function that integrates validation + templates.

**Styles:** `teaching`, `interview`, `storytelling`

````python
# ============================================================
# TASK 6: Prompt Generator App
# ============================================================

# Style-based prompt templates
STYLE_TEMPLATES = {
    "teaching":     PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style, step by step"
    ),
    "interview":    PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Generate 3 interview questions about {topic} for {audience} in a {tone} tone"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} as a story for {audience} in a {tone} tone"
    ),
}


def generate_prompt(topic: str, audience: str, tone: str, style: str) -> str:
    """
    Full Prompt Generator App — validates inputs and generates a dynamic prompt.

    Args:
        topic    (str): The subject to explain.
        audience (str): Target audience (beginner / intermediate / expert).
        tone     (str): Desired tone (formal / casual / fun).
        style    (str): Prompt style (teaching / interview / storytelling).

    Returns:
        str: The fully formatted, dynamic prompt.

    Raises:
        ValueError: If style is not one of the allowed styles.
    """
    # Step 1: Validate audience and tone
    audience, tone = validate_inputs(audience, tone, use_defaults=True)

    # Step 2: Validate style
    style = style.strip().lower()
    if style not in STYLE_TEMPLATES:
        raise ValueError(f"Invalid style '{style}'. Choose from: {list(STYLE_TEMPLATES.keys())}")

    # Step 3: Select the appropriate template and format it
    selected_template = STYLE_TEMPLATES[style]
    return selected_template.format(topic=topic, audience=audience, tone=tone)


# --- Test Cases ---
print("📌 Task 6 Output – Prompt Generator App:\n")

test_cases = [
    ("Neural Networks", "beginners", "fun",    "storytelling"),
    ("Python",          "expert",    "formal",  "interview"),
    ("Data Science",    "intermediate", "casual", "teaching"),
    ("AI",              "novice",    "weird",   "teaching"),   # Invalid → defaults applied
]

for topic, audience, tone, style in test_cases:
    print(f"  Inputs → topic: '{topic}' | audience: '{audience}' | tone: '{tone}' | style: '{style}'")
    prompt = generate_prompt(topic, audience, tone, style)
    print(f"  Output → {prompt}")
    print()
````

## ✅ Task 7: Template Reusability Test (10%)

**Goal:** Demonstrate that **one single template** can produce 5 different outputs — same structure, different content.

````python
# ============================================================
# TASK 7: Template Reusability Test
# ============================================================

# ONE reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# 5 different inputs to run through the same template
five_inputs = [
    {"topic": "Artificial Intelligence", "audience": "beginners",    "tone": "friendly"},
    {"topic": "Python Programming",      "audience": "kids",          "tone": "fun"},
    {"topic": "Deep Learning",           "audience": "engineers",     "tone": "technical"},
    {"topic": "Data Science",            "audience": "intermediate",  "tone": "casual"},
    {"topic": "Cloud Computing",         "audience": "executives",    "tone": "formal"},
]

print("📌 Task 7 Output – Template Reusability Test (1 Template × 5 Inputs):\n")
print(f"  Template Used: '{reusable_template.template}'\n")
print("-" * 65)

for i, inputs in enumerate(five_inputs, start=1):
    output = reusable_template.format(**inputs)
    print(f"  Run {i}: {output}")

print("-" * 65)
print("\n  ✅ Same template structure — 5 unique outputs. Reusability confirmed!")
````

## 🔁 Full Pipeline Demo

**Flow:** `User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output`

````python
# ============================================================
# FULL PIPELINE: End-to-End Prompt Engine Demo
# ============================================================

def run_prompt_pipeline(topic: str, audience: str, tone: str, style: str) -> str:
    """
    Full pipeline: User Input → Validation → Template → Output.

    Args:
        topic    (str): Subject to explain.
        audience (str): Target audience.
        tone     (str): Desired tone.
        style    (str): Prompt style.

    Returns:
        str: Final generated prompt.
    """
    print("  [Step 1] 📥 Raw User Input received")
    print(f"           topic='{topic}', audience='{audience}', tone='{tone}', style='{style}'")

    print("  [Step 2] 🔍 Validating inputs...")
    validated_audience, validated_tone = validate_inputs(audience, tone, use_defaults=True)
    print(f"           audience='{validated_audience}', tone='{validated_tone}' ✅")

    print("  [Step 3] 📋 Selecting PromptTemplate...")
    final_prompt = generate_prompt(topic, validated_audience, validated_tone, style)
    print(f"           Style selected: '{style}' ✅")

    print("  [Step 4] ⚙️  Generating dynamic prompt...")
    print(f"  [Step 5] 📤 OUTPUT → {final_prompt}")
    return final_prompt


# --- Run Pipeline Demo ---
print("📌 Full Pipeline Demo:\n")
print("=" * 65)
result = run_prompt_pipeline(
    topic    = "Neural Networks",
    audience = "beginners",
    tone     = "fun",
    style    = "storytelling"
)
print("=" * 65)
````
